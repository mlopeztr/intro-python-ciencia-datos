**Métodos de cadena**

Podemos aplicar diferentes funciones a las cadenas como haríamos en Excel. Sin embargo, en Python utilizamos métodos. Un método es una función que "pertenece" a un objeto. Para llamar a un método utilizamos el signo "." después del objeto .  Veamos algunos métodos de cadenas para cambiar el caso del texto.

**Variables**

Las variables nos ayudan a almacenar valores de datos. En Python, a menudo trabajamos con datos, por lo que las variables son útiles para gestionar estos datos adecuadamente. Una variable contiene un valor, que es la información asociada a una variable. Para asignar un valor a una variable, usamos el =.

Creemos un mensaje que diga "Estoy aprendiendo Python" y almacenémoslo en una variable llamada mensaje_1.

```python
# mensaje_1 = "Estoy aprendiendo Python"
mensaje_1 = "Estoy aprendiendo Python"
```

Ahora, si queremos obtener el mensaje "Estoy aprendiendo Python" sólo tenemos que escribir el nombre de la variable

```python
mensaje_1
```

Podemos crear tantas variables como queramos. Sólo hay que asegurarse de asignar nombres diferentes a las nuevas variables. Creemos un nuevo mensaje que diga "¡y es divertido!" y guardémoslo en una variable llamada mensaje_2.

```python
# mensaje_2 = "y es divertido!"
mensaje_2 = "y es divertido!"
```

**String Concatenation**

Podemos juntar el mensaje_1 y el mensaje_2 utilizando el operador +. Esto se llama concatenación de cadenas.

```python
mensaje_1 + mensaje_2
```

Necesitamos añadir un espacio en blanco " "para que el texto sea legible. Esta vez también vamos a almacenar este nuevo mensaje en una nueva variable llamada mensaje

```python
mensaje = mensaje_1 + ' ' + mensaje_2
mensaje
```

También podemos utilizar otra opción llamada f-string. Veamos

```python
mensaje = f'{mensaje_1} {mensaje_2}'
mensaje
```

**List**

En Python, las listas se utilizan para almacenar múltiples elementos en una sola variable. Las listas son contenedores ordenados y mutables. En Python, llamamos "mutables" a los objetos que pueden cambiar sus valores. Es decir, los elementos de una lista pueden cambiar su valor.

Para crear una lista, debemos introducir los elementos dentro de corchetes ([]), separados por comas. Consideremos la siguiente lista llamada "países" que contiene algunos de los países con mayor población del mundo.

```python
paises = ['Estados Unidos', 'India', 'China', 'Colombia']
```

Tenga en cuenta que las listas pueden tener elementos de diferentes tipos y elementos duplicados.

**Indexación**



Mediante la indexación podemos obtener un elemento de la lista por su posición. Cada elemento de una lista tiene un índice (posición en la lista). Python utiliza la indexación basada en el cero. Es decir, el primer elemento ("Estados Unidos") tiene un índice 0, el segundo("India") tiene un índice 1, y así sucesivamente.

Para acceder a un elemento por su índice debemos utilizar corchetes []. Veamos algunos ejemplos:

```python
print(paises[0])
print(paises[1])
print(paises[2])
print(paises[3])
```

También existe el índice negativo que nos ayuda a obtener elementos a partir de la última posición de la lista, por lo que en lugar de utilizar los índices de 0 y superiores, utilizamos los índices de -1 e inferiores.

Obtengamos el último elemento de la lista pero ahora usando un índice negativo.

```python
print(paises[-1])
```

**Slicing**

El corte significa acceder a partes de una lista. Una rebanada es un subconjunto de elementos de la lista. La notación de rebanadas tiene la forma de

nombre_de_la_lista[inicio:parada]

donde "inicio" representa el índice del primer elemento, y parada representa el elemento en el que hay que detenerse (sin incluirlo en la rebanada).

Veamos algunos ejemplos:

```python
print(paises[0:3])
print(paises[1:])
print(paises[:2])
```

**Operaciones y métodos de list**

Hay muchas cosas que se pueden hacer con las listas. Puedes añadir nuevos elementos, eliminar cualquier elemento, actualizar valores, etc. En esta sección, veremos las operaciones y métodos más comunes de las listas

**Añadir elementos a una lista**

Hay diferentes métodos que nos ayudan a añadir un nuevo elemento a una lista. Echemos un vistazo.

```python
# append: Añade un nuevo elemento al final de una lista
paises = ['Estados Unidos', 'India', 'China', 'Colombia']
paises.append('Ecuador')
paises
```

```python
# insert: Añade un nuevo elemento en una posición específica
paises = ['Estados Unidos', 'India', 'China', 'Colombia']
paises.insert(0,'Ecuador')
paises
```

```python
# Unir listas con operadores
paises_2 = ['Reino Unido', 'Alemania', 'Austria']
paises + paises_2
```

Podemos poner la lista paises y paises_2 dentro de otra lista, lo que se llama lista anidada.

```python
# nested list
nested_list = [paises, paises_2]
nested_list
```

**Eliminar un elemento**

Existen diferentes métodos que nos ayudan a eliminar un elemento de una lista.

```python
paises = ['Estados Unidos', 'India',
             'China', 'Colombia']
print(paises)

# .remove() Elimina el primer valor coincidente
paises.remove('Estados Unidos')
print(paises)
# .pop() Elimina un elemento en un índice específico y lo devuelve.
paises.pop(0)
print(paises)
# del: elimina un elemento en un índice específico
del paises[0]
print(paises)
```

**Ordenar una lista**

Podemos ordenar fácilmente una lista utilizando el método .sort(). Creemos una nueva lista llamada números y luego ordenémosla del número más pequeño al más grande.

```python
numeros = [4, 3, 10, 7, 1, 2]
numeros.sort()
numeros
```

Podemos añadir el argumento reverse al método .sort() para controlar el orden. Si queremos que sea descendente establecemos reverse=True

```python
numeros = [4, 3, 10, 7, 1, 2]
numeros.sort(reverse=True)
numeros
```

**Actualizar el valor de una lista**

Para actualizar un valor en una lista utilizamos la indexación para localizar el elemento que queremos actualizar y luego lo establecemos con un nuevo valor utilizando el signo =.

```python
numeros[0] = 1000
numeros
```

**Copiar una lista**

Siempre que necesitemos hacer una copia de una lista existente, podemos utilizar las siguientes opciones:

```python
nueva_lista = paises[:]
print(nueva_lista)
nueva_lista_2 = paises.copy()
print(nueva_lista_2)
```

The [:] comes from the cutting technique we learned earlier. In this case, we set neither a start value nor a stop value, so the entire list is selected. On the other hand, the .copy() method explicitly makes a copy of a list.
